# Mempool Transactions

```python
import pandas as pd
import datetime
import time
```

```python
df = pd.read_csv('amount_transactions.csv')
```

```python
timestamps = []
for date in df.date:
    year = int(date[0:4])
    month = int(date[6:7])
    day = int(date[9:10])
    hours = int(date[11:13])
    minutes = int(date[14:16])
    
    t = int(datetime.datetime(year,month,day,hours,minutes).timestamp())
    timestamps.append(t)
    
df['timestamps'] = timestamps
```

```python
df
```

```python
df_blocks = pd.read_csv('tx_per_block.csv')
```

```python
df_blocks[-3:]
```

```python
plt.plot([df_blocks.timestamp.iloc[0], df_blocks.timestamp.iloc[0]],
         [df['amount_tx'].min(), df['amount_tx'].max()], '--', color='grey', label='Block')

for t in df_blocks.timestamp[1:]:
    plt.plot([t, t], [df['amount_tx'].min(), df['amount_tx'].max()], '--', color='grey')

_ = plt.plot([min(timestamps), max(timestamps)] ,[df['amount_tx'].quantile(), df['amount_tx'].quantile()], color='firebrick')
_ = plt.plot([min(timestamps), max(timestamps)] ,[df['amount_tx'].quantile(0.25), df['amount_tx'].quantile(0.25)], '--', color='firebrick')
_ = plt.plot([min(timestamps), max(timestamps)] ,[df['amount_tx'].quantile(0.75), df['amount_tx'].quantile(0.75)], '--',color='firebrick')
_ = plt.plot(timestamps, df['amount_tx'], label='Transaktionen')
_ = plt.title('Transaktionen im Mempool')
_ = plt.legend(loc='upper left')
_ = plt.xticks([])
```

```python
x_vals = [i for i in range(df_blocks['tx_count'].shape[0])]
_ = plt.plot([min(x_vals), max(x_vals)] ,[df_blocks['tx_count'].quantile(), df_blocks['tx_count'].quantile()], color='firebrick')
_ = plt.plot([min(x_vals), max(x_vals)] ,[df_blocks['tx_count'].quantile(0.25), df_blocks['tx_count'].quantile(0.25)], '--', color='firebrick')
_ = plt.plot([min(x_vals), max(x_vals)] ,[df_blocks['tx_count'].quantile(0.75), df_blocks['tx_count'].quantile(0.75)], '--',color='firebrick')
_ = plt.plot(x_vals, df_blocks['tx_count'])
_ = plt.title('Anzahl Transaktionen in einem Block')
_ = plt.xticks([])
```

```python
df_blocks['tx_count'].quantile()
```

# Zeitdifferenz zwischen zwei BlÃ¶cken

```python
dates = [datetime.datetime.fromtimestamp(t) for t in df_blocks.timestamp]
```

```python
differences = []
for i in range(1,len(dates)):
    differences.append((dates[i] - dates[i-1]).seconds)
```

```python
_ = plt.plot([0, len(differences)], [np.quantile(differences, 0.5), np.quantile(differences, 0.5) ], '--', color='firebrick')
_ = plt.plot(differences)
_ = plt.title('Zeitliche Differenz zwischen zwei BlÃ¶cken')
_ = plt.ylabel('Zeit in s')
_ = plt.xlabel('BlÃ¶cke')
_ = plt.xticks([])
```

```python
print(f"Median der Erstellungszeit eines Blocks: {np.quantile(differences, 0.5) }")
print(f"in Minuten: {np.quantile(differences, 0.5) / 60 }")
```

```python
print(f"Alle {np.quantile(differences, 0.5) / 60} Minuten werden {df_blocks['tx_count'].mean()} Transaktionen aus dem Mempool entfernt.")
```

# Transaktionen in 10 Minuten

```python
new_txs = []

for i in range(1, len(df)):
    start = df.iloc[i-1]
    end = df.iloc[i]
    diff = (datetime.datetime.fromtimestamp(end.timestamps) - 
            datetime.datetime.fromtimestamp(start.timestamps)).seconds / 60
    if diff >= 4.0 and diff <= 6.0:
    
    
        block_txs =  df_blocks[(df_blocks.timestamp >= start.timestamps) & 
                           (df_blocks.timestamp <= end.timestamps)].tx_count.sum()

        new_txs.append(end.amount_tx + block_txs - start.amount_tx)
```

```python
len(new_txs)
```

```python
x_vals = [i for i in range(len(new_txs))]
_ = plt.plot([0, len(x_vals)], [np.quantile(new_txs, 0.5), np.quantile(new_txs, 0.5)], '--', color='firebrick')
_ = plt.plot([0, len(x_vals)], [np.mean(new_txs), np.mean(new_txs)], '--', color='deeppink')
_ = plt.plot(x_vals, new_txs)
_ = plt.xticks([])
_ = plt.title("Neue Transaktionen pro 5 Minuten")
```

```python
print(f"Alle 5 Minuten gibt es in etwa durchschnittlich {np.mean(new_txs)} neue Transaktionen")
```

```python
print(f"Alle 5 Minuten gibt es in etwa {np.quantile(new_txs, 0.5)} neue Transaktionen")
```

```python
pred_10 = df_blocks['tx_count'].mean() / np.quantile(differences, 0.5) * 600
```

```python
print(f"Alle zehn Minuten werden {int(pred_10)} Transaktionen in BlÃ¶cke verpackt, wÃ¤hrend {int(np.quantile(new_txs, 0.5)*2)} neue Transaktionen erstehen")
```
